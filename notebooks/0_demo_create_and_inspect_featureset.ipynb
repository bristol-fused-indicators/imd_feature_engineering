```python
from pathlib import Path
from imd_features.config import FeatureSetConfig, GroupConfig, ReductionMethod
from imd_features.process import create_feature_set
from imd_features.diagnostic import (
    group_summary,
    correlation_within_groups,
    correlation_between_groups,
    correlation_full,
    distribution_plot,
)
import polars as pl
from project_paths import paths

input_file = paths.input_file
```

```python

config = FeatureSetConfig(
    name="notebook_demo",
    description="Demo config for inspection",
    groups={
        "crime": GroupConfig(
            columns=[
                "violent-crime", 
                "burglary", 
                "anti-social-behaviour",
                "shoplifting", 
                "criminal-damage-arson", 
                "drugs",
                "total_crimes", 
                "resolution_rate",
            ],
            scale=True,
        ),
        "connectivity": GroupConfig(
            columns=[
                "Overall (walking)", 
                "Overall (cycling)",
                "Overall (public transport)", 
                "Overall (driving)", 
                "Overall",
            ],
            scale=True,
            reduction_method=ReductionMethod.PCA,
            n_components=2,
        ),
    },
)

df, metadata, _ = create_feature_set(input_file, config)
```

```python
group_summary(config, metadata)
```

```python
distribution_plot(df, config)
```

```python
within_corrs = correlation_within_groups(df, config)
for name, fig in within_corrs.items():
    fig.show()
```
